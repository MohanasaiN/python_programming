```````python
"""
write a program to find the student grades based on marks
1.if student marks in between 100 to 80 then grade is A
2.if student marks in between 79 to 60 then grade is B
3.if student marks in between 59 to 40 then grade is C
4.if student marks in between 39 to 25 then grade is D
5.if student marks in between 24 to 0 then grade is Fail
6.if other than above marks expected output should be please enter correct marks!
"""
marks=int(input("enter marks here:"))
if(marks>80 and marks<=100):
    print(marks,"Is A grade")
elif(marks>=60 and marks<=79):
    print(marks,"Is B grade")
elif(marks>=40 and marks<=59):
    print(marks,"Is C grade")
elif(marks>=25 and marks<=39):
    print(marks,"Is D grade")
elif(marks>=0 and marks<=24):
    print(marks,"Is Fail")
else:
    print(marks,"Is not correct marks,please enter correct marks")
```````

```````python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if(a >= b and a >= c):
    print("The largest number is:", a)
elif (b >= a and b >= c):
    print("The largest number is:", b)
else:
    print("The largest number is:", c)
```````

```````python
year = int(input("Enter a year: "))
if year % 400 == 0:
    print(year, "is a Leap Year")
elif year % 100 == 0:
    print(year, "is not a Leap Year")
elif year % 4 == 0:
    print(year, "is a Leap Year")
else:
    print(year, "is not a Leap Year")
```````
